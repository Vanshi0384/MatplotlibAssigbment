```python
#Question 1:

#Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in Python. It offers a variety of functionalities for:

#Generating plots: Line plots, scatter plots, bar charts, histograms, pie charts, 3D plots, and many more.
#Customizing plots: You can control line styles, colors, markers, labels, titles, legends, and other visual elements.
#Embedding in applications: Matplotlib plots can be integrated into web applications, desktop applications, and reports.
#Here are five common plots you can create using the pyplot module of Matplotlib:

#Line Plots: Used to visualize trends or relationships between continuous variables over time or another continuous variable on the x-axis.

#Scatter Plots: Useful for exploring relationships between two independent variables. Each data point is represented by a marker on the plot.

#Bar Charts: Ideal for comparing discrete categories or groups. The height or length of each bar represents the value for that category.

#Histograms: Employed to visualize the distribution of continuous data. It shows the frequency of data points within specific ranges (bins).

#Pie Charts: Represent the proportion of each category or group to the whole. Useful for visualizing categorical data where the parts sum to 100%.
```

```python
#Question 2:
#A scatter plot is a type of graphical representation used to visualize the relationship between two numeric variables. It utilizes a Cartesian coordinate system with:

#X-axis: Represents one variable from your data set.
#Y-axis: Represents another variable from your data set.
#Each data point in your data set is plotted as a marker (like a dot, circle, or other symbol) on the scatter plot. The marker's position on the x and y axes corresponds to the values of the two variables for that particular data point.

#Here's why scatter plots are widely used:

#Identifying Relationships: By observing the distribution of the markers on the plot, you can identify potential relationships between the two variables. These relationships can be:
#Positive correlation: Markers show an upward trend, suggesting that as the value on the x-axis increases, the value on the y-axis also tends to increase.
#Negative correlation: Markers show a downward trend, indicating that as the value on the x-axis increases, the value on the y-axis tends to decrease.
#No correlation: Markers appear randomly scattered, suggesting no clear relationship between the two variables.
#Identifying Outliers: Scatter plots can reveal outliers, which are data points that fall far away from the majority of the other points. These outliers may warrant further investigation.
#Visually Exploring Trends: Scatter plots provide a visual way to explore trends or patterns in your data. This can be particularly helpful in the initial stages of data analysis when you're trying to get a sense of your data.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))

plt.figure(figsize=(8, 6))  # Set figure size for better visualization
plt.scatter(x, y)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
#Question 3:
#The subplot() function in libraries like Matplotlib (Python) is used to create multiple plots within a single figure window. 
#It allows you to efficiently organize and compare different visualizations of your data. 

import matplotlib.pyplot as plt
import numpy as np

# Data for four lines
line1_x = np.array([0, 1, 2, 3, 4, 5])
line1_y = np.array([0, 100, 200, 300, 400, 500])

line2_x = np.array([0, 1, 2, 3, 4, 5])
line2_y = np.array([50, 20, 40, 20, 60, 70])

line3_x = np.array([0, 1, 2, 3, 4, 5])
line3_y = np.array([10, 20, 30, 40, 50, 60])

line4_x = np.array([0, 1, 2, 3, 4, 5])
line4_y = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

# Plot each line on a separate subplot
axes[0, 0].plot(line1_x, line1_y, label='Line 1')
axes[0, 0].set_title('Line 1')
axes[0, 0].legend()

axes[0, 1].plot(line2_x, line2_y, label='Line 2')
axes[0, 1].set_title('Line 2')
axes[0, 1].legend()

axes[1, 0].plot(line3_x, line3_y, label='Line 3')
axes[1, 0].set_title('Line 3')
axes[1, 0].legend()

axes[1, 1].plot(line4_x, line4_y, label='Line 4')
axes[1, 1].set_title('Line 4')
axes[1, 1].legend()

# Adjust layout
fig.suptitle('Four Line Plots using subplot()')
plt.tight_layout()

plt.show()
```

```python
#Question 4:

#A bar plot is a type of graphical representation that uses bars to depict categories and their associated values. 
#It is a common and versatile tool for visualizing data comparisons between different groups or entities.


import matplotlib.pyplot as plt
import numpy as np

# Company names and profit
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(company, profit, color=['red', 'green', 'blue', 'purple'])
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company Profit (Bar Plot)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(company, profit, color=['red', 'green', 'blue', 'purple'])
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Company Profit (Horizontal Bar Plot)")
plt.gca().invert_yaxis()  # Invert y-axis to display highest profit at the top
plt.tight_layout()
plt.show()
```

```python
# Question 5:
#A box plot is a visual representation of a dataset's distribution, summarizing key statistics in a single plot. Here's why box plots are used:

# Distribution Summary: They provide a quick overview of the data's spread, central tendency, and potential outliers.
#Identifying Outliers: Box plots use whiskers to extend from the box and indicate outliers, which are data points that fall outside a specific range from the rest of the data.
#Comparison of Multiple Datasets: By placing multiple box plots side-by-side, you can easily compare the distribution of different datasets and identify significant differences in measures like central tendency or spread.


import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate your data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot of two datasets
sns.boxplot(
    data=[box1, box2],
    showmeans=True,  # Show the mean as a point within the box
    notch=True,  # Add notches to the box for better spread visualization
)

# Customize the plot (optional)
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.title("Box Plot Comparison")
plt.show()
```
